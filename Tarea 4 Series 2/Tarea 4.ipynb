# B)

Según el criterio de la raiz al numerro de la muestras, debe ser menor o igual a 11 ,dicho lo anterior con respecto al grafico y a la relación entre años y meses de estaciones mas bien frias tomaremos $p_1=7$, $p_2=5$ , con  d=12 por ser valores anuales y fijaremos $R_1=(-\infty,0.28)$ y $R_2=(0.28,\infty)$ dado el grafico tambien.

```r
model_tar<-tar(flu, 7, 5, 12, is.constant1 = TRUE, is.constant2 = TRUE, transform = "no",
 center = FALSE, standard = FALSE, estimate.thd = FALSE, threshold=0.28, 
method = c("MAIC", "CLS")[1],  order.select = TRUE, print = TRUE)
```

Dados los valores p del estadistico F ademas del AIC bajo, podemos decir que el modelo es adecuado.

# C)

```r
a=predict(model_tar,n.ahead=4 ,n.sim=1000)
a$fit
a$pred.interval
```
